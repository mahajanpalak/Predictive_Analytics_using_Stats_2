```python
# Import Required Libraries
import numpy as np
import pandas as pd
```

```python
# Load the Dataset
df = pd.read_csv("data.csv", encoding="latin1", engine="python")
print(df.columns)
```

```python
x = df['no2'].values
```

```python
#  roll no.
r = 102497010

# Compute Parameters
a_r = 0.05 * (r % 7)
b_r = 0.3 * (r % 5 + 1)
print("a_r =", a_r)
print("b_r =", b_r)
```

```python
# Removing NA values
df['no2'].isna().sum()
df_no2 = df[['no2']].dropna()
x = df_no2['no2'].values
```

```python
# Apply the Transformation
z = x + a_r * np.sin(b_r * x)
print(z)
```

Screenshot 2026-02-04 145253.png

```python
# Estimated mean
mu = np.mean(z)
print("Estimated mu =", mu)
```

```python
# Estimate Variance
sigma_sq = np.mean((z - mu) ** 2)
print("Estimated sigma^2 =", sigma_sq)
```

```python
# Compute lambda
lam = 1 / (2 * sigma_sq)
print("Estimated lambda =", lam)
```

```python
# Compute Normalization constant C
c = 1 / np.sqrt(2 * np.pi * sigma_sq)
print("Estimated c =", c)
```

```python
print("Final parameters:")
print("lambda =", lam)
print("mu =", mu)
print("c =", c)
```
